```python
import pandas as pd
data_frame = pd.read_csv('data.csv')
data_frame 
```

```python
data_frame.values
```

```python
df1 = pd.DataFrame(data_frame, columns=['Name', 'Club', 'Best Overall Rating'])
df1 = df1.dropna(subset=['Name', 'Best Overall Rating'])
df1
df1_top = df1.nlargest(20, 'Best Overall Rating')
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(12, 6))
plt.scatter(df1_top['Name'], df1_top['Best Overall Rating'])

# Labels and title
plt.xlabel('Name')
plt.ylabel('Best Overall Rating')
plt.title('Player Name vs Best Overall Rating')

plt.xticks(rotation=90)  # rotate names so they don't overlap
plt.tight_layout()
plt.show()
```
